```python
import pandas as pd
import numpy as np

# mostrando os dados para identificar quais linhas e colunas precisam ser tradadas
df = pd.read_excel("Telco_customer_churn.xlsx")
df
```

```python
# encontrando valores nan no dataframe

nan_df = df.isna()
print(nan_df)
```

```python
# os unicos valores nan, são sobre os motivos de churn.
# estabelecendo que os valores nan representam pessoas que não tiveram motivos
## substituindo valores por "No opinion" (Sem opinião)
df_limpo = df.copy()
df_limpo["Churn Reason"] = df_limpo["Churn Reason"].fillna("No opinion")
print(df_limpo)

```
